# Carga de Datos a través de la función OPEN

```python
data3 = open(mainpath + "/" + "customer-churn-model/Customer Churn Model.txt","r") #r:read , w:escribir , a: adicionar
```

```python
cols = data3.readline().strip().split(",") 
#next navega por cada una de las lineas 
#strip elimina espacios en blanco al inicio y al final de la linea
#split divide en fragmentos la linea de acuerdo al idetificador deseado
cols
```

```python
n_cols  = len(cols) #len numero de columnas
```

```python
n_cols
```

```python
counter = 0
main_dic = {}
for col in cols:
    main_dic[col] =[]
```

```python
for line in data3:
    value=line.strip().split(",")
    for i in range(len(cols)):
        main_dic[cols[i]].append(value[i])
    counter += 1
    
print("El data set tiene %d filas y %d columnas" %(counter,n_cols))
```

```python
df3 = pd.DataFrame(main_dic)
df3.head()
```

# Lectura y escritura de ficheros

```python
infile = mainpath + "/" + "customer-churn-model/Customer Churn Model.txt"
outfile = mainpath + "/" + "customer-churn-model/Tab Customer Churn Model.txt"
```

```python
with open(infile,"r") as infile1:
    with open(outfile,"w") as outfile1:
        for line in infile1:
            fields= line.strip().split(",")
            outfile1.write("\t".join(fields))
            outfile1.write("\n")
```

```python
df4 = pd.read_csv(outfile, sep="\t")
df4.head()
```

# Leer Datos desde una URL

```python
medals_url = "http://winterolympicsmedals.com/medals.csv"
```

```python
medals_data = pd.read_csv(medals_url)
```

```python
medals_data.head()
```

```python
import csv #gestiona ficheros csv
import urllib3 #navegar y acceder informacion desde una URL
```

```python
http = urllib3.PoolManager()
r= http.request("GET",medals_url)
r.status
response = r.data
response
```

```python
cr= csv.reader(response)
for row in cr:
    print(row)
```

```python
for line in cr:
    value = line.strip().split("\n")
```
