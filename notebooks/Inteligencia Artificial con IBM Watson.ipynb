- Empecemos, ¬øQu√© es REST?

REST (Representational State Transfer) es un estilo de arquitectura para aplicaciones con comunicaci√≥n/conexi√≥n a la red, la utilizamos para poder comunicarnos entre m√°quinas utilizando HTTP

- ¬øQu√© es HTTP?

Hyper Text Transfer Protocol es un protocolo que utiliza la World Wide Web que define la forma en la que los mensajes se deben de estructurar y mandar, as√≠ como sus respuestas.

Las aplicaciones que utilizan REST usan peticiones HTTP para crear, alterar, leer y borrar datos. En resumen REST utiliza HTTP para operaciones CRUD (Create, Read, Update, Delete).

Al usar m√©todos HTTP usamos varios verbos para la interacci√≥n, algunos de los m√°s importantes son:

- GET : Com√∫nmente cuando solo queremos obtener algo. Ej. cargar una p√°gina web

- PUT: Cuando queremos modificar
