# Tipos de Apredizaje

 El **Aprendizaje Estadístico** trata de cómo aprender de conjuntos de datos: cómo predecir y estimar cantidades importantes o qué patrones interesantes se pueden descubrir en ellos.



El **aprendizaje supervisado** es una técnica para deducir una función a partir de datos de entrenamiento. Los datos de entrenamiento consisten de pares de objetos: una componente del par son los datos de entrada y el otro, los resultados deseados.Este algoritmo aprenderá de este anlisis y emitirá una respuesta. Este aprendizaje requiere de la supervisión de una persona. Esta persona supervisara el procesamiento de estos datos.



**Aprendizaje no supervisado** es un método de Aprendizaje Automático donde un modelo se ajusta a las observaciones. Se distingue del Aprendizaje supervisado por el hecho de que no hay un conocimiento a priori. En el aprendizaje no supervisado, un conjunto de datos de objetos de entrada es tratado.El algoritmo aprende a interpretar los datos y con ello a construir decisiones.



**Aprendizaje por refuerzo** es un área del aprendizaje automático inspirada en la psicología conductista, cuya ocupación es determinar qué acciones debe escoger un agente de software en un entorno dado con el fin de maximizar alguna noción de “recompensa” o premio acumulado. El algoritmo va aprendiendo de manera cíclica, el algoritmo interpreta y escoge el guarismo más adecuado para la situación, se elige la mejor acción y esta es “premiada”. De esta manera el algoritmo puede aprender cual es la mejor acción para determinado contexto. Este sistema aprende a base de ensayo y error.

# Álgebra lineal

El álgebra lineal puede ser utilizada para mejorar el código y hacer que los algoritmos de Machine Learning sean más rápidos

Encontrar el vector que al multiplicarlo por una matriz da un resultado.





El resultado al tratar de resolver el sistemas de ecuaciones nos puede dar:
    


### Sistema Compatible Determinado

Los 3 planos (ecuaciones) coinciden en un punto.



### Sistema Compatible Indeterminado

Los 3 planos (ecuaciones) no coinciden en un punto, puede ser una recta o un plano, es decir, las soluciones pueden ser infinitas.



### Sistema Incompatible 

No se tiene una solucion. PUede existir una interseccion para 2 ecuaciones pero no para la tercera, o que no coincidan para nada entre ellas.



```python
#Se definen las matrices.
matrizA = np.matrix([
                    [1,-3,2],
                    [5,6,-1],
                    [4,-1,3]
                    ])
matriz_resultado = np.array([
                        [-3],
                        [13],
                        [8]
                        ])
display(matrizA);display(matriz_resultado)
```

```python
#Se encuentra la inversa de la matriz
matriz_inversa = np.linalg.inv(matrizA)
```

```python
#Se multiplica la inversa de la matriz por el resultado

matriz_solucion = matriz_inversa.dot(matriz_resultado) #matriz_solucion = matriz_inversa * matriz_resultado


matriz_solucion
```

```python
#Se comprueba el resultado
matrizA * matriz_solucion
```

En matemáticas, la gráfica de una función es un tipo de representación gráfica que permite conocer intuitivamente el comportamiento de dicha función, si es continua, si es infinita, si corta, etc. Más formalmente dada una función: el gráfico es el conjunto de todos los pares ordenados de la función f, es decir, como un subconjunto del producto cartesiano X×Y.

La tabla de valores nos permite graficar de unaforma facil y sencilla en uun plano la funcion, no es metodo mas optimo pero nos puede servir para una primera impresion

# Límites, Infinito y asíntotas. Teoremas

### Limites

Para la matemática, un límite es una magnitud a la que se acercan progresivamente los términos de una secuencia infinita de magnitudes. Un límite matemático, por lo tanto, expresa la tendencia de una función o de una sucesión mientras sus parámetros se aproximan a un cierto valor.

Se parte de una funcion y el limite lo que pretende es evaluar lo que realiza dicha funcion apartir unicamente de su expresion. Evalua la funcion en un punto dado.





Ejemplo:

### Asintota

Las asíntotas verticales son rectas verticales a las cuales la función se va acercando indefinidamente sin llegar nunca a cortarlas (discontinuidad de la funcion).



Lo que se prentende con este tipo de funciones es  evaluar los limites laterales para determinar su comportamiento.




### Sucesión 

Es una representacion de numeros. Se puede aplicar limites para conocer si la sucecion diverge (tiende a infinito) o converge (tiende a un numero real)

Una sucesión es divergente cuando no tiene límite. Es decir, cuando no existe ningún número finito al cual se aproxima. Cuando una sucesión es divergente, decimos que no existe su límite:

Una sucesión es convergente cuando tiene límite finito. El límite de una sucesión es el número al que la sucesión se aproxima cada vez más.

# Teoremas

**Teorema de Bolzano**

Si una función f(x) es continua en un intervalo [a,b], cuyos valores extremos f(a) y f(b) tienen distinto signo, la función cortará al eje (y por tanto devolverá el valor 0) en algún punto del intervalo.







**Teorema de los Valores Intermedios**

En el mismo caso y condiciones explicitadas en el teorema de Bolzano, la función alcanzará cualquier valor intermedio comprendido entre f(a) y f(b)]. Como puedes ver, no es más que una generalización del teorema de Bolzano.
