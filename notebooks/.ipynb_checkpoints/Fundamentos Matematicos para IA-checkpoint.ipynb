# Tipos de Apredizaje

 El **Aprendizaje Estadstico** trata de cmo aprender de conjuntos de datos: cmo predecir y estimar cantidades importantes o qu patrones interesantes se pueden descubrir en ellos.



El **aprendizaje supervisado** es una tcnica para deducir una funcin a partir de datos de entrenamiento. Los datos de entrenamiento consisten de pares de objetos: una componente del par son los datos de entrada y el otro, los resultados deseados.Este algoritmo aprender de este anlisis y emitir una respuesta. Este aprendizaje requiere de la supervisin de una persona. Esta persona supervisara el procesamiento de estos datos.



**Aprendizaje no supervisado** es un mtodo de Aprendizaje Automtico donde un modelo se ajusta a las observaciones. Se distingue del Aprendizaje supervisado por el hecho de que no hay un conocimiento a priori. En el aprendizaje no supervisado, un conjunto de datos de objetos de entrada es tratado.El algoritmo aprende a interpretar los datos y con ello a construir decisiones.



**Aprendizaje por refuerzo** es un rea del aprendizaje automtico inspirada en la psicologa conductista, cuya ocupacin es determinar qu acciones debe escoger un agente de software en un entorno dado con el fin de maximizar alguna nocin de recompensa o premio acumulado. El algoritmo va aprendiendo de manera cclica, el algoritmo interpreta y escoge el guarismo ms adecuado para la situacin, se elige la mejor accin y esta es premiada. De esta manera el algoritmo puede aprender cual es la mejor accin para determinado contexto. Este sistema aprende a base de ensayo y error.
